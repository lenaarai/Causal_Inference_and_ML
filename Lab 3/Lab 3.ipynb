## Import Data

```python
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}
html = requests.get(
    "https://stats.nba.com/stats/leagueLeaders?LeagueID=00&PerMode=Totals&Scope=S&Season=2017-18&SeasonType=Regular+Season&StatCategory=FG3_PCT",
    headers=headers
).text
```

## Functions to clean data

```python
def get_columns_from(html):
    return ast.literal_eval(html[html.find('headers')+9:html.find('headers')+203])


def get_data_from(html):
    return_list = []
    for row in ast.literal_eval(html[html.find('rowSet')+9:len(html)-3]):
        return_list.append(tuple(row))
    return return_list


df = pd.DataFrame.from_records(get_data_from(html),
columns=get_columns_from(html))
```
